```python
# ============================================
# 1. Retail Business Flow
# ============================================

# 1. Customer visits a store or website
# 2. Customer places an order
# 3. Order contains one or more products
# 4. Payment is completed
# 5. Order is shipped or delivered
# 6. Sales data is analyzed for business insights
```

```python
# ============================================
# 2. Why Data is Important in Retail
# ============================================

# Retail businesses analyze data to:
# - Track daily / monthly sales
# - Identify top-selling products
# - Understand customer behavior
# - Optimize inventory
# - Increase revenue
```

```python
# ============================================
# 3. Analytics Types in Retail
# ============================================

# Descriptive Analytics:
#   - What happened? (Daily sales, revenue)
#
# Diagnostic Analytics:
#   - Why did it happen? (Low sales in a store)
#
# Predictive Analytics:
#   - What will happen? (Sales forecasting)
#
# Prescriptive Analytics:
#   - What should we do? (Discount strategies)
```
